```python
class Person:
    name:str = ''

    # initializer
    def __init__(self, name): 
        self.name = name
        self.say_hello()

    def set_name(self, name):
        self.name = name

    def get_name(self):
        return self.name
    
    def get_family_name(self):
        return self.name[:1]
    
    def get_personal_name(self):
        return self.name[1:]

    def say_hello(self):
        print(f'Hello! I am {self.name}')

    def say_bye(self, name):
        print(f'Goodbye! {name}')

    def get_self_id(self):
        print(f'Class Object ID : {id(self)}')

# self : 클래스 인스턴스. 인스턴스를 가르킨다. id(self) 로 메모리 위치 확인 가능.
# parameter(매개변수) := variable(변수). 메소드 선언의 변수 목록
# argument(인수, 인자) := value(값). 메소드가 호출될 때 전달되는 실제 값

# 생성자 constructor __new__ & __init__
person1, person2 = Person('Yang'), Person('Shin')
print(person1.name)
print(person2.name)

# Class 확인 == 객체의 데이터 타입
print(f'person1: {type(person1)}')
print(f'person2: {type(person2)}')

# 객체의 함수 호출
person1.say_hello() 
person2.say_hello()

person1.say_bye('Kim') 
person2.say_bye('Yang')

# Setter (?)
person1.set_name('Kim')
person2.set_name('Yang')
print(f'person1 name: {person1.name}')
print(f'person2 name: {person2.name}')


person1.say_hello() 
person2.say_hello()

# getter setter
person1.set_name('김철수')

print(person1.get_name())
print(person1.get_family_name())
print(person1.get_personal_name())
```

```python
class LogicGate:
    def __init__(self, w1, w2, b):
        self.w1 = w1
        self.w2 = w2
        self.b = b

    # __call__ : 함수를 지정하지 않고 호출함. 그러므로 하나만 설정 가능
    # self(x1, x2) -> self(x1, x2) 로 사용하게 됨.
    def __call__(self, x1, x2):
        z = self.w1 * x1 + self.w2 * x2 +self.b
        if z > 0: return 1
        else: return 0




class ANDGate:
    def __init__(self):
        self.gate = LogicGate(0.5, 0.5, -0.7)

    def __call__(self, x1, x2):
        return self.gate(x1, x2)
    
class NANDGate:
    def __init__(self):
        self.gate = LogicGate(-0.5, -0.5, 0.7)

    def __call__(self, x1, x2):
        return self.gate(x1, x2)

class ORGate:
    def __init__(self):
        self.gate = LogicGate(0.5, 0.5, -0.2)

    def __call__(self, x1, x2):
        return self.gate(x1, x2)
    
class NORGate:
    def __init__(self):
        self.gate = LogicGate(-0.5, -0.5, 0.2)

    def __call__(self, x1, x2):
        return self.gate(x1, x2)
    
class XORGate:
    def __init__(self):
        self.or_ = ORGate()
        self.nand_ = NANDGate()
        self.and_ = ANDGate()
    
    def __call__(self, x1, x2):
        p = self.or_(x1, x2)
        q = self.nand_(x1, x2)
        return self.and_(p, q)

class XNORGate:
    def __init__(self):
        self.nor_ = NORGate()
        self.and_ = ANDGate()
        self.or_ = ORGate()

    def __call__(self, x1, x2):
        p = self.nor_(x1, x2)
        q = self.and_(x1, x2)
        return self.or_(p, q)


if __name__ == '__main__':
    # 클래스 내부에서 검증 코드 작성시 저장된 변수가 실제 사용에 영향을 줄수 있다.
    # __name__ 메서드를 검증 용도로 사용 가능.
    # 패키지 사용할 때 아래 코드는 실행 되지 않는다.
    TT = [1, 1]
    FT = [0, 1]
    TF = [1, 0]
    FF = [0, 0]

    gate = XNORGate()
    print(gate(*TT))
    print(gate(*TF))
    print(gate(*FT))
    print(gate(*FF))

```
