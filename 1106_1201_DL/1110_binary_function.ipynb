```python
# Binary Cross Entropy
import numpy as np
class AffineFunction:
    def __init__(self, w:list, b:float):
        self.w = w
        self.b = b

    def __call__(self, x):
        z = np.dot(self.w, x) + self.b
        return z
    
class Sigmoid:
    def __call__(self, z):
        a = 1 / (1 + np.exp(-z))
        return a

class BCELoss:
    def __init__(self):
        super().__init__()

    def entropy(self, x1, x2):
        return np.log(x2)*x1

    def forward(self, y, pred):
        j = -(self.entropy(y, pred) + self.entropy((1 - y), (1 - pred)))
        return j

loss_function = BCELoss()
print(loss_function.forward(0, 0.111))
preds = np.arange(0.1, 1, 0.1)


print('Case.1 y = 0')
for pred in preds:
    J = loss_function.forward(0, pred)
    print(f'{pred = :.1f} -> {J = :.4f}')
    
print('\nCase.2 y = 1')
for pred in preds:
    J = loss_function.forward(1, pred)
    print(f'{pred = :.1f} -> {J = :.4f}')



import matplotlib.pyplot as plt
bce1 = loss_function.forward(1, preds)
bce2 = loss_function.forward(0, preds)
plt.plot(preds, bce1)
plt.plot(preds, bce2)
plt.show()
```
