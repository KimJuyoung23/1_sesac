# Artificial Neural Network : 인공 신경망
x -> Affine Function -> Activation Function(Sigmoid Function) -> y

```python
import numpy as np

class AffineFunction:
    def __init__(self, w:list, b:float):
        self.w = w
        self.b = b

    def __call__(self, x):
        z = np.dot(self.w, x) + self.b
        return z
    
class Sigmoid:
    def __call__(self, z):
        a = 1 / (1 + np.exp(-z))
        return a

class ArtificialNeural:
    def __init__(self, w:list, b:float):
        self.affine = AffineFunction(w, b)
        self.activation = Sigmoid()

    def __call__(self, x):
        z = self.affine(x)
        a = self.activation(z)
        return a # affine value

class Model:
    def __init__(self):
        self.AND = ArtificialNeural([5, 5], -7.5)
        self.NAND = ArtificialNeural([-5, -5], 7.5)
        self.OR = ArtificialNeural([5, 5], -2.5)

    def __call__(self, x:list):
        a1 = round(self.AND(x), 4)
        a2 = round(self.NAND(x), 4)
        a3 = round(self.OR(x), 4)
        a = np.array([a1, a2, a3])
        return a
    
```

뉴런 : neuron. (ArtificialNeural 클래스로 만든 하나의 객체)
뉴런 구성 : weight vector(w1, w2, ... ), scalar bias
레이어 : 뉴런으로 구성된 

```python
TT = [1, 1]
FT = [0, 1]
TF = [1, 0]
FF = [0, 0]


model = Model()
print(model(FF))
print(model(TF))
print(model(FT))
print(model(TT))
```

```python
# Affine Function
w = [1, 1]
b = -1.5
affine1 = AffineFunction(w, b)
x = [1, 1]
print(affine1(x))

# Sigmoid
sigmoid = Sigmoid()
print(f'z=-5: {sigmoid(-5)}')
print(f'z=-3: {sigmoid(-3)}')
print(f'z=0: {sigmoid(0)}')
print(f'z=3: {sigmoid(3)}')
print(f'z=5: {sigmoid(5)}')

# ArtificialNeuron
artificial_neural = ArtificialNeural(w, b)
print(artificial_neural(x))

# Sigmoid Graphics
import matplotlib.pyplot as plt
x = np.linspace(-5, 5, 100)
y = sigmoid(x)
plt.axvline(0, 0, 1, color='gray', linestyle='-', linewidth=1)
plt.axhline(0, 0, 1, color='gray', linestyle='-', linewidth=1)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
plt.show()
```
