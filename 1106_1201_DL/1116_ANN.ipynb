```python
import matplotlib.pyplot as plt
import numpy as np

spot = np.linspace(-0.5, 1.5, 10)
x1, x2 = np.meshgrid(spot, spot)
x2 = x2[::-1]
x1, x2 = x1.flatten(), x2.flatten()

# w1, b1 = [-0.5, -0.5], 0.75
# w2, b2 = [0.5, -0.5], 0.25
# w3, b3 = [0, -0.5], 0.25
# w4, b4 = [0.5, 0.5], -0.75

w1, b1 = [-0.5], 0.75
w2, b2 = [0.5], 0.25
w3, b3 = [0], 0.25


def affine(x1, x2, w:list, b):
    y = []
    for x in x1:
        z = np.dot(w, x) + b
        y.append(z)
    return np.array(y)
L1_m1 = affine(x1, x2, w1, b1)
L1_m2 = affine(x1, x2, w2, b2)
L1_m3 = affine(x1, x2, w3, b3)
# L2_m4 = affine(X, w4, b4)


plt.plot(x1, L1_m1)
plt.plot(x1, L1_m2)
plt.plot(x1, L1_m3)

```
