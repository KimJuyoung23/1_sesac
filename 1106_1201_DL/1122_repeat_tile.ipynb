```python
# check pattern image
import numpy as np
import matplotlib.pyplot as plt

white_patch = 255 * np.ones(shape=(10, 10))
black_patch = 0 * np.ones(shape=(10, 10))

img1 = np.hstack([white_patch, black_patch])
img2 = np.hstack([black_patch, white_patch])
img = np.vstack([img1, img2])

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')

ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
# check pattern image
import numpy as np
import matplotlib.pyplot as plt

white_patch = 255 * np.ones(shape=(10, 10))
black_patch = 0 * np.ones(shape=(10, 10))

img1 = np.hstack([white_patch, black_patch, white_patch])
img2 = np.hstack([black_patch, white_patch, black_patch])
img = np.vstack([img1, img2, img1])

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')

ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
# np.repeat & np.tile
import numpy as np

data = np.arange(5)
print(data)

# np.repeat -> 원소별 반복
print(f'repeat: {np.repeat(data, repeats=3)}')

# np.tile -> 전체 패턴 반복
print(f'tile  : {np.tile(data, reps=3)}')
```

```python
import numpy as np
data = np.arange(6).reshape(2, 3)
print(data)

print(f'repeat(axis=0):\n{np.repeat(data, repeats=3, axis=0)}')
# print(data.repeat(repeats=3, axis=0))

print(f'repeat(axis=1):\n{np.repeat(data, repeats=3, axis=1)}')
# print(data.repeat(repeats=3, axis=1))

print(f'repeat(axis=0 and axis=1):\n{np.repeat(np.repeat(data, repeats=2, axis=0), repeats=3, axis=1)}')
```

```python
import numpy as np
data = np.arange(6).reshape(2, 3)
print(data)

print(f'tile(axis=0):\n{np.tile(data, reps=[3,1])}')
print(f'tile(axis=1):\n{np.tile(data, reps=[1,3])}')
print(f'tile(axis=0 and axis=1):\n{np.tile(data, reps=[3,3])}')
```

```python
import numpy as np

white_patch = 255 * np.ones(shape=(10, 10))
black_patch = 0 * np.ones(shape=(10, 10))

img1 = np.hstack([white_patch, black_patch])
img2 = np.hstack([black_patch, white_patch])
img = np.vstack([img1, img2])

img = np.tile(img, reps=[4, 4])

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
# check pattern image w/ np.tile
import numpy as np
import matplotlib.pyplot as plt

patch1_1 = 255 * np.ones(shape=(10, 10))
patch1_2 = 150 * np.ones(shape=(10, 10))
patch2_1 = 150 * np.ones(shape=(10, 10))
patch2_2 = 50 * np.ones(shape=(10, 10))
patch1 = np.hstack([patch1_1, patch1_2])
patch2 = np.hstack([patch2_1, patch2_2])
patch = np.vstack([patch1, patch2])

img = np.tile(patch, reps=[4,4])

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
# grayscale gradation image w/ np.repeat
import numpy as np
import matplotlib.pyplot as plt

img = np.arange(0, 256, 50).reshape(1, -1)
img = img.repeat(100, axis=0).repeat(30, axis=1)
                                     
fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
# grayscale gradation image w/ np.repeat
img = np.arange(0, 151, 50).reshape(1, -2)
img = img.repeat(100, axis=0).repeat(30, axis=1)

fig, axes = plt.subplots(1,2,figsize=(7,7))
axes[0].imshow(img, cmap='gray')
axes[1].imshow(img, cmap='gray', vmax=255, vmin=0)

axes[0].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
axes[1].tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
img = np.arange(0, 256, 50)[::-1].reshape(-1, 1)
img = img.repeat(30, axis=0).repeat(100, axis=1)

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```

```python
interval = 1
img1 = np.arange(0, 256, interval).reshape(1, -1)
img1 = img1.repeat(300, axis=0).repeat(2, axis=1)
img2 = np.arange(0, 256, interval)[::-1].reshape(1, -1)
img2 = img2.repeat(300, axis=0).repeat(2, axis=1)

img = np.vstack([img1, img2])

fig, ax = plt.subplots(figsize=(7,7))
ax.imshow(img, cmap='gray')
ax.tick_params(left=False, labelleft=False, bottom=False, labelbottom=False)
```
