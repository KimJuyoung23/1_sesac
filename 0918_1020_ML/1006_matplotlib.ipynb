```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator
import seaborn as sns

figsize = (7, 7)
```

```python
n_student = 100
math_scores = np.random.normal(loc=50, scale=10, size=(100, ))

medianprops = {'linewidth': 5, 'color' : 'k'} # k == black, b == blue
boxprops = {'linestyle' : '--', 'color' : 'k', 'alpha' : 0.7}
whiskerprops = {'linestyle' : '--', 'color' : 'tab:blue', 'alpha' : 0.8}


fig, ax = plt.subplots(figsize=figsize)
ax.boxplot(math_scores, # data
           notch=True, # 신뢰 구간을 노치로 표현
           whis=3, # whiskers. 수염의 위치
           sym='bx', # 
           showfliers=False, # outliers 표시 유무 # tykey style
           vert=False,
           medianprops=medianprops, # median 중앙값
           boxprops=boxprops, # box
           whiskerprops=whiskerprops, # wishkers 수염, 꼬리
           capprops=whiskerprops # 최대 최소
           ) # True : 세로, False : 가로
```

```python
n_student = 100
math_scores = np.random.normal(50, 15, size =(100, 1))
chem_scores = np.random.normal(70, 10, size =(n_student, 1))
phy_scores = np.random.normal(30, 12, size =(n_student, 1))
pro_scores = np.random.normal(80, 5, size =(n_student, 1))
labels = ['math', 'chem', 'phy', 'pro']

data = np.hstack((math_scores, chem_scores, phy_scores, pro_scores))
whiskerprops = {'linestyle' : '--', 'color' : 'tab:blue', 'alpha' : 0.8}
medianprops = {'linewidth': 3, 'color' : 'tab:red'}

plt.style.use('seaborn')
fig, ax = plt.subplots(figsize = figsize)
ax.set_ylim(0, 100)
ax.boxplot(data,
           labels=labels,
           notch=True,
           medianprops=medianprops,
           whiskerprops=whiskerprops,
           capprops=whiskerprops, # whisker 와 cap을 동일하게 표시
           )

ax.yaxis.set_major_locator(MultipleLocator(20)) # 20 단위
ax.yaxis.set_minor_locator(MultipleLocator(5))  # 5 단위

ax.grid(axis='y', # y axis major ticks 넓이 조절
        linewidth=2,
        )
ax.grid(axis='y', # y axis minor ticks 점선
        which='minor',
        linewidth=2,
        linestyle=':')
ax.grid(axis='x', # x axis 라인 제거
        linewidth=0)
```

```python
n_student = 100
math_scores = np.random.normal(50, 15, size =(100, 1))
chem_scores = np.random.normal(70, 10, size =(n_student, 1))
phy_scores = np.random.normal(30, 12, size =(n_student, 1))
pro_scores = np.random.normal(80, 5, size =(n_student, 1))
labels = ['math', 'chem', 'phy', 'pro']

data = np.hstack((math_scores, chem_scores, phy_scores, pro_scores))
whiskerprops = {'linestyle' : '--', 'color' : 'tab:blue', 'alpha' : 0.8}
medianprops = {'linewidth': 3, 'color' : 'tab:red'}

plt.style.use('seaborn')
fig, axes = plt.subplots(2, 1, figsize = figsize)
axes[0].set_ylim(0, 100)
axes[0].boxplot(data,
           labels=labels,
           notch=True,
           medianprops=medianprops,
           whiskerprops=whiskerprops,
           capprops=whiskerprops, # whisker 와 cap을 동일하게 표시
           )
axes[1].violinplot(data)

axes[0].yaxis.set_major_locator(MultipleLocator(20)) # 20 단위
axes[0].yaxis.set_minor_locator(MultipleLocator(5))  # 5 단위

axes[0].grid(axis='y', # y axis major ticks 넓이 조절
        linewidth=2,
        )
axes[0].grid(axis='y', # y axis minor ticks 점선
        which='minor',
        linewidth=2,
        linestyle=':')
axes[0].grid(axis='x', # x axis 라인 제거
        linewidth=0)

```
