
Example 2-3. 키와 몸무게를 입력 받아, 비만도(BMI)를 판별하는 프로그램 만들기
                        (*아래의 표를 참고하여 작성 )

image.png

```python
height = float(input('키를 입력해주세요.(cm)'))
weight = float(input('몸무게를 입력해주세요.(kg)'))
meter_height = height/100
BMI = weight / (meter_height * meter_height)

if BMI < 18.5:
  print('저체중')
elif BMI < 22.9:
  print('정상')
elif BMI < 24.9:
  print('과체중')
elif BMI < 29.9:
  print('경도비만')
else:
  print('고도비만')
```

# 반복문

## while문

```python
# 나무를 열번 찍고 열번 찍으면 나무가 넘어간다.
treeHit = 0
while treeHit < 10:
  treeHit = treeHit + 1
  print(f'나무를 {treeHit}번 찍었습니다.')
  if treeHit == 10:
    print('나무 넘어갑니다~~~~~')
```

```python
# break 문 활용하기(반복문을 빠져 나온다.)
coffee = 10
money = 500
while money:
  print('돈을 받았으니 커피를 줍니다.')
  coffee = coffee - 1
  print(f'남은 커피의 양은 {coffee}개 입니다.')
  if not coffee: # coffee가 False
    break
```

```python
# continue : 현재 반복을 중지하고 다음 반복으로 넘가는 역할
a = 0
while a < 10:
  a = a + 1
  if a % 2 == 0:
    continue
  print(a)
```

## 연습문제

Example 2-4. A 부서의 월 매출이 아래와 같을 때, For문을 이용하여, A 부서의 평균 월 매출 계산
(* 리스트와 For문을 이용해서 계산 / 1월 : 5500 / 2월 : 5300 / 3월 : 6700 / 4월 : 5500 / 5월 : 5800)



```python
sales = [5500, 5300, 6700, 5500, 5800]
total_sales = 0

for sale in sales:
  total_sales = total_sales + sale
print(total_sales/len(sales))
```

Example 2-5. 반복문을 활용하여 아래와 같이 * 기호를 피라미드로 출력하는 프로그램 제작




```python
for i in range(5):
    for j in range(5):
        if i >= j:
            print('*', end='')
    print('')
```

Example 2-7. For문과 If문을 이용하여, 정수를 입력 받아 정수의 약수를 구하는 프로그램 제작

	     예)  입력 : 15  / 결과 : 1,3,5,15






```python
num = int(input('숫자를 입력해주세요.'))
result = []
for i in range(1, num+1):
  if num % i == 0:
    result.append(i)
print(result)
```

(과제)Example 2-8. For문과 If문을 이용하여, 아래의 리스트의 최댓값을 구하는 프로그램 제작

	list = [1,24,3,44,1000,2,0,74]     / 결과 : 1000

```python
list1 = [1,24,3,44,1000,2,0,74]
result = 0
for i in list1:
  if i > result:
    result = i
```
