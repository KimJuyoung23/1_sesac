## 전역변수 vs 지역변수
- 전역변수 : 함수 외부에서 선언 된 변수이며 프로그램 전체에서 사용 가능하다.
- 지역변수 : 함수 내부에서 선언 된 변수이며 해당 함수 내부에서만 사용할 수 있다.

```python
def test():
  local_var = 20
test()
```

```python
print(local_var)
```

```python
def test2():
  local_var = 20
  return local_var
a = test2()
print(a)
```

```python
def test3():
  global local_var
  local_var = 20
test3()
print(local_var)
```

```python
local_var = 20
```

## 연습문제

Example 2-9. 입력하는 모든 수의 평균을 계산하는 함수 만들기
* 입력 개수에 상관없이 사용하기 위해, 함수 입력 값 자리에, *args를 사용
->  def means(*args):

```python
def means(*args):
  sum1 = 0
  count1 = 0
  for i in args:
    sum1 = sum1 + i
    count1 = count1 + 1
  result = sum1/count1
  return result
print(means(1,2,3,4,5,6,7,8))
```

```python
def my_means(*args):
    rst = sum(args) / len(args)
    return rst

my_means(1,2,3)
```


Example 2-10. 두 수를 입력 받아 사칙연산(+,-, x, /)을 하는 함수 만들기

```python
def cal(num1, num2):
  return num1*num2, num1-num2, num1*num2, num1/num2
```

```python
cal()
```

```python
def cul2(num1, num2):
    a = input("기호를 입력하시오 : ")
    if a == '+' :
        return num1 + num2
    elif a == '-' :
        return num1 - num2
    elif a == '*' :
        return num1 * num2
    elif a == '/' :
        return num1 / num2
    elif a == '//' :
        return num1 // num2
    elif a == '%' :
        return num1 % num2
    else :
        return '제공되지 않는 기호 입니다.'
cul2(10, 5)
```
