### 집합

```python
s1 = set([1,2,3])
s2 = {2,3,4,4}
print(s1)
print(s2)
```

```python
s1.add(4)
print(s1)
s2.update([5,6,7])
print(s2)
s2.remove(7)
print(s2)
```

```python
print(s1)
print(s2)
print(s1&s2) # 교집합 1
print(s1.intersection(s2)) # 교집합 2
print(s1|s2) # 합집합 1
print(s1.union(s2)) # 합집합 2
print(s1-s2) # 차집합 1
print(s1.difference(s2)) # 차집합 2
```
