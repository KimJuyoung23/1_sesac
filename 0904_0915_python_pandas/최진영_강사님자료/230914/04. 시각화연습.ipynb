```python
import pandas as pd
df = pd.read_csv('shopping_mall.csv', index_col = 0)
df['price'] = df['item_price'] * df['quantity']
```

```python
q1 = df[['pref', 'price']].groupby('pref', as_index =False).sum().sort_values('price', ascending = False)
```

```python
import plotly.express as px
fig = px.bar(q1, x='pref', y='price', color = 'pref')
fig.show()
```

```python
fig = px.pie(q1, values='price', names='pref')
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

```python
# 성별 상품 선호도?
q2 = df[['gender', 'quantity', 'item_name']]
q2g = q2.groupby(['item_name', 'gender'], as_index = False).sum()
```

```python
fig = px.bar(q2g, x="item_name", y="quantity", color="gender", barmode="group")
fig.show()
```
