```python
import pandas as pd
mpg = pd.read_csv('https://raw.githubusercontent.com/jin0choi1216/dataset/main/mpg.csv')
mpg.head()
```

2. 자동차 배기량과 고속도로 연비 시각화
3. 자동차 구동방식 별 도시연비 비교 시각화

```python
# 1
q1 = mpg[['manufacturer', 'model']].groupby('manufacturer').count().sort_values('model', ascending=False)
len(q1['model'])
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots(figsize = (15, 5))
bar_colors = ['tab:red', 'tab:red', 'tab:red', 'tab:red', 'tab:orange','tab:orange','tab:orange','tab:orange','tab:orange',
              'tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue']

ax.bar(q1.index , q1['model'].values,color = bar_colors)
plt.show()
```

```python
import matplotlib.pyplot as plt


fig, ax = plt.subplots()
bar_colors = ['tab:red', 'tab:red', 'tab:red', 'tab:red', 'tab:orange','tab:orange','tab:orange','tab:orange','tab:orange',
              'tab:blue','tab:blue','tab:blue','tab:blue','tab:blue','tab:blue']

ax.barh(q1.index , q1['model'].values,color = bar_colors)
plt.show()
```

```python
# 자동차 구동방식 별 도시연비 비교 시각화
q3 = mpg[['drv', 'cty']].groupby('drv').mean()
plt.bar(q3.index, q3['cty'])
```
